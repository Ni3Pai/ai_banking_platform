```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas

def create_pdf(filename, title, content):
    c = canvas.Canvas(filename, pagesize=A4)
    width, height = A4

    # Title
    c.setFont("Helvetica-Bold", 14)
    c.drawString(50, height-50, title)

    # Content
    c.setFont("Helvetica", 9)
    y = height - 80
    for line in content:
        if y < 50:
            c.showPage()
            y = height - 50
        c.drawString(50, y, line)
        y -= 14

    c.save()
    print(f"вњ… Created: {filename}")

# Document 1 вЂ” ESG Policy
create_pdf("esg_policy.pdf", "NORTHERN EUROPE BANK вЂ” ESG REPORTING POLICY 2024", [
    "Section 1 вЂ” Scope",
    "This policy applies to all corporate banking clients with annual turnover exceeding EUR 10 million.",
    "All clients must report ESG metrics annually.",
    "",
    "Section 2 вЂ” Reporting Requirements",
    "Clients must report the following metrics:",
    "- GHG Scope 1 and Scope 2 emissions in tonnes CO2 equivalent",
    "- Energy consumption in MWh",
    "- Water usage in cubic metres",
    "- Percentage of renewable energy used",
    "",
    "Section 3 вЂ” EU Taxonomy Alignment",
    "All clients in high-impact sectors must assess activities against EU Taxonomy criteria.",
    "DNSH assessment is mandatory for High Risk sectors including Energy, Manufacturing and Transport.",
    "",
    "Section 4 вЂ” Deadlines",
    "Annual ESG reports must be submitted by 31 March each year.",
    "Non-compliant clients will be flagged for relationship review.",
    "",
    "Section 5 вЂ” Consequences of Non-Compliance",
    "Clients failing to submit reports face:",
    "- Credit facility review",
    "- Potential loan covenant breach",
    "- Escalation to senior relationship management",
])

# Document 2 вЂ” Credit Risk Policy
create_pdf("credit_risk_policy.pdf", "NORTHERN EUROPE BANK вЂ” CREDIT RISK POLICY 2024", [
    "Section 1 вЂ” Credit Risk Framework",
    "All corporate lending decisions must follow the bank credit risk framework.",
    "Minimum credit score required for new facilities is 650.",
    "",
    "Section 2 вЂ” Risk Classification",
    "Clients are classified into three risk tiers:",
    "- Green: Low risk, score above 750, standard terms apply",
    "- Amber: Medium risk, score 650-750, enhanced monitoring required",
    "- Red: High risk, score below 650, credit committee approval required",
    "",
    "Section 3 вЂ” Exposure Limits",
    "Single client exposure must not exceed 10% of total portfolio.",
    "Sector concentration limit is 25% of total portfolio.",
    "Real estate sector limit is 30% due to collateral availability.",
    "",
    "Section 4 вЂ” Review Frequency",
    "Green clients: Annual review",
    "Amber clients: Semi-annual review",
    "Red clients: Quarterly review and enhanced due diligence",
    "",
    "Section 5 вЂ” Reporting",
    "Credit risk reports must be submitted to the Risk Committee monthly.",
    "Exceptions above EUR 5 million require immediate escalation.",
])

# Document 3 вЂ” AML Policy
create_pdf("aml_policy.pdf", "NORTHERN EUROPE BANK вЂ” AML COMPLIANCE POLICY 2024", [
    "Section 1 вЂ” Purpose",
    "This policy establishes Anti-Money Laundering controls for all banking activities.",
    "All staff must complete annual AML training.",
    "",
    "Section 2 вЂ” Customer Due Diligence",
    "Standard CDD required for all new clients.",
    "Enhanced Due Diligence required for:",
    "- Politically Exposed Persons (PEPs)",
    "- High risk jurisdictions",
    "- Complex ownership structures",
    "- Clients with turnover above EUR 50 million",
    "",
    "Section 3 вЂ” Transaction Monitoring",
    "All transactions above EUR 10,000 must be monitored.",
    "Suspicious transactions must be reported within 24 hours.",
    "Cash transactions above EUR 5,000 require additional documentation.",
    "",
    "Section 4 вЂ” Reporting Obligations",
    "Suspicious Activity Reports must be filed with FIU Netherlands.",
    "Annual AML compliance report due by 28 February each year.",
    "Sanctions screening must be performed daily.",
    "",
    "Section 5 вЂ” Penalties",
    "Non-compliance with AML regulations may result in:",
    "- Regulatory fines up to EUR 10 million",
    "- Criminal prosecution of responsible officers",
    "- Revocation of banking license",
])

print("\nвњ… All 3 documents created!")
```

```python
from langchain_groq import ChatGroq
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from google.colab import userdata
import os

os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

llm = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=50)

print("вњ… LangChain ready")
```

```python
# Load all PDFs
pdf_files = ["esg_policy.pdf", "credit_risk_policy.pdf", "aml_policy.pdf"]

all_documents = []

for pdf in pdf_files:
    loader = PyPDFLoader(pdf)
    docs = loader.load()

    # Tag each chunk with source document
    for doc in docs:
        doc.metadata["source"] = pdf

    all_documents.extend(docs)
    print(f"вњ… Loaded: {pdf} вЂ” {len(docs)} pages")

print(f"\nрџ“љ Total documents loaded: {len(all_documents)}")
```

```python
# Split all documents
all_chunks = splitter.split_documents(all_documents)
print(f"вњ… Total chunks: {len(all_chunks)}")

# Create ONE vector store from all documents
vectorstore = FAISS.from_documents(all_chunks, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

print("вњ… Combined vector store created")
print("рџ”Ќ Will search across all 3 documents simultaneously")
```

```python
prompt = ChatPromptTemplate.from_template("""
You are a banking compliance expert at Northern Europe Bank.
Answer questions using ONLY the provided policy documents.
Always mention which document/section your answer comes from.
If information spans multiple documents mention all of them.

Policy Documents:
{context}

Question: {question}

Answer (with document citations):""")

def format_docs_with_source(docs):
    formatted = []
    for doc in docs:
        source = doc.metadata.get("source", "Unknown")
        formatted.append(f"[{source}]\n{doc.page_content}")
    return "\n\n".join(formatted)

rag_chain = (
    {"context": retriever | format_docs_with_source,
     "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)

print("вњ… Multi-document RAG chain ready")
```

```python
questions = [
    "What are all the reporting deadlines across our policies?",
    "What happens to clients who don't comply with our policies?",
    "What triggers enhanced due diligence or enhanced monitoring?",
    "What are the exposure and transaction limits?",
]

print("рџ“љ MULTI-DOCUMENT Q&A вЂ” Northern Europe Bank Policies")
print("=" * 60)

for question in questions:
    print(f"\nвќ“ {question}")
    answer = rag_chain.invoke(question)
    print(f"рџ’¬ {answer}")
    print("-" * 60)
```
